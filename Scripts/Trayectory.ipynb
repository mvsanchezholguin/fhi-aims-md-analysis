```python
import re

input_file = "salida.out"          # salida de FHI-aims
output_file = "trajectory.xyz"

atoms_block = []
frames = []

with open(input_file, "r") as f:
    lines = f.readlines()

i = 0
while i < len(lines):
    line = lines[i]

    # Detectar inicio de bloque de coordenadas
    if "Atomic structure (and velocities)" in line:
        atoms_block = []
        i += 1

        while i < len(lines):
            line = lines[i]

            # Línea tipo: atom   x   y   z   Element
            if line.strip().startswith("atom"):
                parts = line.split()
                x, y, z = parts[1:4]
                element = parts[4]
                atoms_block.append((element, x, y, z))

            # Fin del bloque
            if line.strip().startswith("-----"):
                break

            i += 1

        if atoms_block:
            frames.append(atoms_block)

    i += 1


# Escribir archivo XYZ
with open(output_file, "w") as out:
    for step, frame in enumerate(frames):
        out.write(f"{len(frame)}\n")
        out.write(f"Step {step}\n")
        for atom in frame:
            out.write(f"{atom[0]} {atom[1]} {atom[2]} {atom[3]}\n")

print(f"✔ Trayectoria generada: {output_file}")
print(f"✔ Número de frames: {len(frames)}")
```
